```python
# Import library
from google_play_scraper import reviews_all, Sort
import csv
import pandas as pd
```

```python
from google_play_scraper import app, reviews_all, Sort

# Scrapping Data Ulasan Google Play
scrapreview = reviews_all(
    'com.mobile.legends', # ID aplikasi
    lang='en', # Bahasa ulasan
    country='id', # Negara
    sort=Sort.MOST_RELEVANT, # Urutan ulasan
    count=11000 # Jumlah maksimal ulasan
)
```

```python
# Simpan dalam CSV
with open('review_game_mobile_legends.csv', mode='w', newline='', encoding='utf-8') as file :
  writer = csv.writer(file)
  writer.writerow(['review'])
  for review in scrapreview :
    writer.writerow([review['content']])
```
